# Load Orcas

```
orcas = pd.read_csv('/content/drive/My Drive/Project 3/orcas.tsv', sep='\t', header=None, encoding='utf-8')
orcas.rename(columns={0:'qid', 1:'query', 2:'did', 3:'url'}, inplace=True)
orcas.dropna(inplace=True)
orcas.reset_index(drop=True, inplace=True)
orcas['url'] = orcas['url'].apply(lambda x: unicodedata.normalize("NFC", x))
orcas['query'] = orcas['query'].apply(lambda x: unicodedata.normalize("NFC", x))
```

```
df_temp = orcas[orcas['qid'].isin(candidates['qid'].unique())]
```

# ICR

```
# Click data
ClickData = orcas['did'].value_counts()
```

```
icr = dict()
for (movie, values) in candidates.groupby('movie'):
  print(movie)
  icr[movie] = list()
  for id, row in values.iterrows():
    qid = row['qid']
    # print(qid)
    giao_did = row['did'].split(",")
    # print(did)
    # tinh tu
    tu = 0
    for d in giao_did:
      tu += ClickData.loc[d]
    # print(tu)
    # tinh mau
    mau = 0
    allDidOfQid = df_temp[df_temp['qid'] == qid]['did'].unique()
    for d in allDidOfQid:
      mau += ClickData.loc[d]
    # print(mau)
    # icr of qid
    icr_qid = tu/mau
    # print(icr_qid)
    if icr_qid > 1:
      print("   " + row['query'])

    icr[movie].append([qid, icr_qid])
```
