```
# clean
noise = ['...', 'Q:', 'Edit', 'See more', 'See more »']
clean_content = dict()
for movie in contentTopLink:
  print(movie)
  doc = []
  text  = contentTopLink[movie]
  for t in " ".join(text).split("\n"):
    t = t.replace(u'\xa0', ' ')
    temp = t.strip()
    temp = temp.replace("|", '')
    if len(temp) > 1 and temp not in noise:
      # print(temp.strip()) 
      doc.append(temp)
  clean_content[movie] = doc
```
